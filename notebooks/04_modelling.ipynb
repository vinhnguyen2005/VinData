## Mục tiêu (Phần 3 của đề thi — 20 điểm)
Train và đánh giá mô hình dự báo cột **Revenue** cho giai đoạn 01/01/2023–01/07/2024.

## Chỉ số đánh giá
- **MAE** — Mean Absolute Error (càng thấp càng tốt)
- **RMSE** — Root Mean Squared Error (càng thấp càng tốt)
- **R²** — Coefficient of Determination (càng cao càng tốt, lý tưởng gần 1)

## Các mô hình cần thử
- Baseline: naive (dùng giá trị ngày hôm qua)
- Statistical: SARIMA, Holt-Winters
- ML: XGBoost, LightGBM, Random Forest
- Deep Learning (optional): LSTM, N-BEATS

## Validation strategy
Time-series cross-validation (Walk-forward) — KHÔNG dùng random split.
