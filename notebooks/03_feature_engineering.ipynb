# 03 — Feature Engineering cho Mô hình Dự báo

## Mục tiêu
Tạo các feature từ `sales.csv` và các bảng phụ trợ để đưa vào mô hình dự báo doanh thu.

## Các nhóm feature cần tạo
- **Lag features**: doanh thu t-1, t-7, t-14, t-30
- **Rolling stats**: rolling mean/std 7 ngày, 30 ngày
- **Calendar features**: day_of_week, month, quarter, is_weekend, is_holiday_VN
- **Trend**: linear trend, seasonal decomposition residual
- **External signals**: web_traffic sessions lag, inventory fill_rate lag
- **Promotion flags**: có khuyến mãi trong ngày không (join promotions)

## ⚠️ Lưu ý chống Data Leakage
Chỉ dùng thông tin có sẵn tại thời điểm t để dự báo t+1.
